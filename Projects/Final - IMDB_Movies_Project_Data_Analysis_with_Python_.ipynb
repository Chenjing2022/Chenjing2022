**3.2: Classify movies based on Ratings [Excellent, Good, and Average] by using applying functions**

In this section we will use apply() function to convert movies with great than 8.0 to a string value of "Excellent" and with great than 6.0 to a string value of "Good" the rest to "Average" and use this transformed values to create a new column.

frist we will need to create a function that, when given a rating which determines if it's Excellent, Good, or Average. 

Secondly, we want to send the entire rating column through this function, which is what apply() does. The .apply() method passes every value in the rating column through the rating_function and then returns a new Series. This Series is then assigned to a new column called Rating_cat.

Lastly, we use value_counts to find out how many movies calssify based on these conditions. 

```python
# Step 1: to create a function that, when given a rating which determines if it's Excellent, Good, or Average 

def rating(rating): 
  if rating > 8.0: 
    return "Excellent"
  elif rating > 6.5:
    return "Good"
  else: 
    return "Average"

# Step 2: to send the entire rating column through this function which is what apply() does 
movies_df['Rating_cat'] = movies_df['Rating'].apply(rating)
movies_df.head() # to display the first 5 rows of dataframe 

Rating_cat_counts =  movies_df["Rating_cat"].value_counts() # to count how many movies classify based on Excellent, Good, and Average.
Rating_cat_counts # to print the count 
```

So, as we can see that there are 59 movies are "Excellent", 544 movies are "Good" and 397 movies are "Average".

 **3.5: Scatterplot of Revenue(Millions) vs Rating**



Does Rating Affect The Revenue?



Next we will plot a scatter plot of the relationship between Ratings and Revenue to find out if the rating has an impact on revenue. In other words, the scatter plot of rating and revenue will help us understand if movies with better rating, generate higher revenue.



```python
def show_scatterplot(movies_drop,var1,var2):
  x = movies_drop[var1]
  y = movies_drop[var2]
  plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
  plt.scatter(x,y, c=y, cmap='Spectral')
  plt.scatter
  plt.xlabel(var1)
  plt.ylabel(var2)

  regression = linregress(x,y) 
  slope = regression.slope
  intercept = regression.intercept
  plt.plot(x, slope*x + intercept)
  plt.show()
```

```python
# Revenue vs Rating 
plt.title("Revenue (Millions) vs Rating")
show_scatterplot(movies_drop,"Rating","Revenue (Millions)")
```

The above scatter plot illustrates although the correlation coefficient is positive but the relationship is very weak.

**3.9: What is the year of the highest average revenue?** 

Here I am going to apply *df.groupby* methods to find the highest average reveune of all movies through the years, by doing so, i will groupby the "Year" column and "Revenue (Millions)"column and take the average of these two column value's mean.  


```python
# barchart to visualise the highest revenue of all movies through out of the year
def show_lineplot(movies_drop):
  
  data = movies_drop.groupby('Year')['Revenue (Millions)'].mean()
  #data
  lables = ["2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016"]
  #lables

  chart = sns.barplot(x=lables, y=data,)
  chart.set_xlabel("Year")
  chart.set_ylabel("'Revenue (Millions)'")
  chart.set_title("Revenue (Millions) by Year")
  chart

show_lineplot(movies_drop)
```

The above chart shows us the average revenue in 2009 was considerably the higher than other years, followed by year 2012. The year 2016 has at least average revenue.
